# Let's play ball!

The object of this exercise is to create the best 'moneyball' baseball team from a given year. My team is from 1985. A moneyball team is made up of players with cheap salaries and high OBP's. I created a list of players that were active in 1985 with at least 10 AB's and then created a ratio of their OBP to their salary. The nine players who played unique positions with the highest ratio make up my team. 

1) Chris Brown 3B
2) Joe Orsulak CF
3) Glenn Davis RF
4) Mark Salas  C
5) Larry Sheets 1B
6) Mariano Duncab 2B
7) Ozzie Guillen SS
8) Tom Browning P
9) Randy Bush LF

```python
del fielding_data['stint']
del fielding_data['teamID']
del fielding_data['G']
del fielding_data['GS']
del fielding_data['InnOuts']
del fielding_data['PO']
del fielding_data['A']
del fielding_data['E']
del fielding_data['DP']
del fielding_data['PB']
del fielding_data['WP']
del fielding_data['SB']
del fielding_data['CS']
del fielding_data['ZR']
```

```python
fielding_data = fielding_data[fielding_data.yearID == 1985]
full_data = pd.merge(fielding_data, all_data)
```

```python
full_data = full_data.drop_duplicates()
```

```python
full_data['OBP'] = (full_data['H']+full_data['BB']+full_data['HBP'])/(full_data['AB']+full_data['BB']+full_data['HBP']+full_data['SF'])

```

```python
full_data.head()
```

```python
full_data['ratio'] = full_data['OBP']/full_data['salary']
```

```python
full_data = full_data.dropna()
```

```python
moneyball85 = full_data.sort(['ratio'], ascending=False)
```

```python
moneyball85 = moneyball85[moneyball85.AB > 10]
```

```python
moneyball85 = moneyball85.reset_index()
```
